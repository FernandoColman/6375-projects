```python
# Make sure to add all the datasets into the runtime 

# Imports
import numpy as np
import pandas as pd
from sklearn import tree
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
from sklearn import model_selection
from sklearn.ensemble import BaggingClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
```

```python
# Functions to handle dataset changes

def mix_sets(valid, train):
  both_arrs = np.vstack([valid, train])
  np.random.shuffle(both_arrs)
  return both_arrs
```

```python
# Functions to tune parameters

def driver(dtc, X_valid, y_valid, X_test, y_test):
    dtc.fit(X_valid, y_valid)
    y_pred = dtc.predict(X_test)
    acc = accuracy_score(y_test,y_pred)*100
    # print("Accuracy: ", acc)
    # print("F1 Score: ", f1_score(y_test, y_pred))
    return acc    
```

```python
def decisionTree(test, train, valid):
    ## print("\nDecision Tree Classifier")
    # Split Target Variable

    X_test = test[:,0:-1]
    y_test = test[:,-1]

    X_train = train[:,0:-1]
    y_train = train[:,-1]

    X_valid = valid[:,0:-1]
    y_valid = valid[:,-1]


    # Tune Parameters using validation set

    hold_acc = {}

    dtc = tree.DecisionTreeClassifier(criterion = "gini", splitter = "best", max_depth = 100)
    hold_acc.update({driver(dtc, X_valid, y_valid, X_test, y_test): dtc})

    dtc = tree.DecisionTreeClassifier(criterion = "gini", splitter = "best", max_depth = 1000)
    hold_acc.update({driver(dtc, X_valid, y_valid, X_test, y_test): dtc})

    dtc = tree.DecisionTreeClassifier(criterion = "gini", splitter = "best", max_depth = 10000)
    hold_acc.update({driver(dtc, X_valid, y_valid, X_test, y_test): dtc})

    dtc = tree.DecisionTreeClassifier(criterion = "gini", splitter = "random", max_depth = 100)
    hold_acc.update({driver(dtc, X_valid, y_valid, X_test, y_test): dtc})

    dtc = tree.DecisionTreeClassifier(criterion = "gini", splitter = "random", max_depth = 1000)
    hold_acc.update({driver(dtc, X_valid, y_valid, X_test, y_test): dtc})

    dtc = tree.DecisionTreeClassifier(criterion = "gini", splitter = "random", max_depth = 10000)
    hold_acc.update({driver(dtc, X_valid, y_valid, X_test, y_test): dtc})

    dtc = tree.DecisionTreeClassifier(criterion = "entropy", splitter = "best", max_depth = 100)
    hold_acc.update({driver(dtc, X_valid, y_valid, X_test, y_test): dtc})

    dtc = tree.DecisionTreeClassifier(criterion = "entropy", splitter = "best", max_depth = 1000)
    hold_acc.update({driver(dtc, X_valid, y_valid, X_test, y_test): dtc})
    dtc = tree.DecisionTreeClassifier(criterion = "entropy", splitter = "best", max_depth = 10000)
    hold_acc.update({driver(dtc, X_valid, y_valid, X_test, y_test): dtc})

    dtc = tree.DecisionTreeClassifier(criterion = "entropy", splitter = "random", max_depth = 100)
    hold_acc.update({driver(dtc, X_valid, y_valid, X_test, y_test): dtc})

    dtc = tree.DecisionTreeClassifier(criterion = "entropy", splitter = "random", max_depth = 1000)
    hold_acc.update({driver(dtc, X_valid, y_valid, X_test, y_test): dtc})

    dtc = tree.DecisionTreeClassifier(criterion = "entropy", splitter = "random", max_depth = 10000)
    hold_acc.update({driver(dtc, X_valid, y_valid, X_test, y_test): dtc})

    dtc = tree.DecisionTreeClassifier(criterion = "log_loss", splitter = "best", max_depth = 100)
    hold_acc.update({driver(dtc, X_valid, y_valid, X_test, y_test): dtc})

    dtc = tree.DecisionTreeClassifier(criterion = "log_loss", splitter = "best", max_depth = 1000)
    hold_acc.update({driver(dtc, X_valid, y_valid, X_test, y_test): dtc})

    dtc = tree.DecisionTreeClassifier(criterion = "log_loss", splitter = "best", max_depth = 10000)
    hold_acc.update({driver(dtc, X_valid, y_valid, X_test, y_test): dtc})

    dtc = tree.DecisionTreeClassifier(criterion = "log_loss", splitter = "random", max_depth = 100)
    hold_acc.update({driver(dtc, X_valid, y_valid, X_test, y_test): dtc})

    dtc = tree.DecisionTreeClassifier(criterion = "log_loss", splitter = "random", max_depth = 1000)
    hold_acc.update({driver(dtc, X_valid, y_valid, X_test, y_test): dtc})

    dtc = tree.DecisionTreeClassifier(criterion = "log_loss", splitter = "random", max_depth = 10000)
    hold_acc.update({driver(dtc, X_valid, y_valid, X_test, y_test): dtc})

    #print("{Accuracies: Model}")
    #display(hold_acc)

    max_key = max(hold_acc)
    ## print("\nThe Model with the highest accuracy is", hold_acc.get(max_key), "with an accuracy of:", max_key)

    # Mix Training and Validation Sets Together
    mixed = mix_sets(valid, train)

    # Split target variable
    X_mixed = mixed[:,0:-1]
    y_mixed = mixed[:,-1]

    # Make Decision Tree Classifier with Best Parameter Settings
    final_dtc = hold_acc.get(max_key)
    final_dtc.fit(X_mixed, y_mixed)
    ## print("Using Best Parameter Settings Found After Tuning and Mixed Valid/Train Sets: ")
    y_pred = final_dtc.predict(X_test)
    acc = accuracy_score(y_test,y_pred)*100
    f1 = f1_score(y_test, y_pred)

    ## print("Accuracy: ", acc)
    ## print("F1 Score: ", f1)
    return acc, f1
```

```python
def baggingClassifier(test, train, valid):
    ## print("\nBagging Classifier")
    # Split Target Variable

    X_test = test[:,0:-1]
    y_test = test[:,-1]

    X_train = train[:,0:-1]
    y_train = train[:,-1]

    X_valid = valid[:,0:-1]
    y_valid = valid[:,-1]

    # Tune Parameters using validation set

    hold_acc = {}

    dtc = BaggingClassifier(max_samples = 1)
    hold_acc.update({driver(dtc, X_valid, y_valid, X_test, y_test): dtc})

    dtc = BaggingClassifier(max_samples = 2)
    hold_acc.update({driver(dtc, X_valid, y_valid, X_test, y_test): dtc})

    dtc = BaggingClassifier(max_samples = 3)
    hold_acc.update({driver(dtc, X_valid, y_valid, X_test, y_test): dtc})

    dtc = BaggingClassifier(max_samples = 4)
    hold_acc.update({driver(dtc, X_valid, y_valid, X_test, y_test): dtc})

    dtc = BaggingClassifier(max_samples = 5)
    hold_acc.update({driver(dtc, X_valid, y_valid, X_test, y_test): dtc})

    dtc = BaggingClassifier(max_samples = 6)
    hold_acc.update({driver(dtc, X_valid, y_valid, X_test, y_test): dtc})

    #print("{Accuracies: Model}")
    #display(hold_acc)

    max_key = max(hold_acc)
    ## print("\nThe Model with the highest accuracy is", hold_acc.get(max_key), "with an accuracy of:", max_key)

    # Mix Training and Validation Sets Together
    mixed = mix_sets(valid, train)

    # Split target variable
    X_mixed = mixed[:,0:-1]
    y_mixed = mixed[:,-1]

    final_dtc = hold_acc.get(max_key)
    final_dtc.fit(X_mixed, y_mixed)
    ## print("Using Best Parameter Settings Found After Tuning and Mixed Valid/Train Sets: ")
    y_pred = final_dtc.predict(X_test)
    acc = accuracy_score(y_test,y_pred)*100
    f1 = f1_score(y_test, y_pred)
    
    ## print("Accuracy: ", acc)
    ## print("F1 Score: ", f1)
    return acc, f1
```

```python
def randomForestClassifier(test, train, valid):
    ## print("\nRandom Forest Classifier")
    # Split Target Variable

    X_test = test[:,0:-1]
    y_test = test[:,-1]

    X_train = train[:,0:-1]
    y_train = train[:,-1]

    X_valid = valid[:,0:-1]
    y_valid = valid[:,-1]

    # Tune Parameters using validation set

    hold_acc = {}

    dtc = RandomForestClassifier(n_estimators = 1)
    hold_acc.update({driver(dtc, X_valid, y_valid, X_test, y_test): dtc})

    dtc = RandomForestClassifier(n_estimators = 2)
    hold_acc.update({driver(dtc, X_valid, y_valid, X_test, y_test): dtc})

    dtc = RandomForestClassifier(n_estimators = 3)
    hold_acc.update({driver(dtc, X_valid, y_valid, X_test, y_test): dtc})

    dtc = RandomForestClassifier(n_estimators = 4)
    hold_acc.update({driver(dtc, X_valid, y_valid, X_test, y_test): dtc})

    dtc = RandomForestClassifier(n_estimators = 5)
    hold_acc.update({driver(dtc, X_valid, y_valid, X_test, y_test): dtc})

    dtc = RandomForestClassifier(n_estimators = 6)
    hold_acc.update({driver(dtc, X_valid, y_valid, X_test, y_test): dtc})

    dtc = RandomForestClassifier(n_estimators = 1, max_depth = 1)
    hold_acc.update({driver(dtc, X_valid, y_valid, X_test, y_test): dtc})

    dtc = RandomForestClassifier(n_estimators = 2, max_depth = 2)
    hold_acc.update({driver(dtc, X_valid, y_valid, X_test, y_test): dtc})

    dtc = RandomForestClassifier(n_estimators = 3, max_depth = 3)
    hold_acc.update({driver(dtc, X_valid, y_valid, X_test, y_test): dtc})

    dtc = RandomForestClassifier(n_estimators = 4, max_depth = 4)
    hold_acc.update({driver(dtc, X_valid, y_valid, X_test, y_test): dtc})

    dtc = RandomForestClassifier(n_estimators = 5, max_depth = 5)
    hold_acc.update({driver(dtc, X_valid, y_valid, X_test, y_test): dtc})

    dtc = RandomForestClassifier(n_estimators = 6, max_depth = 6)
    hold_acc.update({driver(dtc, X_valid, y_valid, X_test, y_test): dtc})

    #print("{Accuracies: Model}")
    #display(hold_acc)

    max_key = max(hold_acc)
    ## print("\nThe Model with the highest accuracy is", hold_acc.get(max_key), "with an accuracy of:", max_key)

    # Mix Training and Validation Sets Together
    mixed = mix_sets(valid, train)

    # Split target variable
    X_mixed = mixed[:,0:-1]
    y_mixed = mixed[:,-1]

    
    final_dtc = hold_acc.get(max_key)
    final_dtc.fit(X_mixed, y_mixed)
    ## print("Using Best Parameter Settings Found After Tuning and Mixed Valid/Train Sets: ")
    y_pred = final_dtc.predict(X_test)
    acc = accuracy_score(y_test,y_pred)*100
    f1 = f1_score(y_test, y_pred)
    
    ## print("Accuracy: ", acc)
    ## print("F1 Score: ", f1)
    return acc, f1
```

```python
def gradientBoostingClassifier(test, train, valid):
    ## print("\nGradient Boosting Classifier")
    # Split Target Variable

    X_test = test[:,0:-1]
    y_test = test[:,-1]

    X_train = train[:,0:-1]
    y_train = train[:,-1]

    X_valid = valid[:,0:-1]
    y_valid = valid[:,-1]

    # Tune Parameters using validation set

    hold_acc = {}

    dtc = GradientBoostingClassifier(loss = "log_loss", learning_rate = 0.1)
    hold_acc.update({driver(dtc, X_valid, y_valid, X_test, y_test): dtc})

    dtc = GradientBoostingClassifier(loss = "log_loss", learning_rate = 1)
    hold_acc.update({driver(dtc, X_valid, y_valid, X_test, y_test): dtc})

    dtc = GradientBoostingClassifier(loss = "log_loss", learning_rate = 10)
    hold_acc.update({driver(dtc, X_valid, y_valid, X_test, y_test): dtc})

    dtc = GradientBoostingClassifier(loss = "exponential", learning_rate = 0.1)
    hold_acc.update({driver(dtc, X_valid, y_valid, X_test, y_test): dtc})

    dtc = GradientBoostingClassifier(loss = "exponential", learning_rate = 1)
    hold_acc.update({driver(dtc, X_valid, y_valid, X_test, y_test): dtc})

    dtc = GradientBoostingClassifier(loss = "exponential", learning_rate = 10)
    hold_acc.update({driver(dtc, X_valid, y_valid, X_test, y_test): dtc})

    #print("{Accuracies: Model}")
    #display(hold_acc)

    max_key = max(hold_acc)
    ## print("\nThe Model with the highest accuracy is", hold_acc.get(max_key), "with an accuracy of:", max_key)

    # Mix Training and Validation Sets Together
    mixed = mix_sets(valid, train)

    # Split target variable
    X_mixed = mixed[:,0:-1]
    y_mixed = mixed[:,-1]

    
    final_dtc = hold_acc.get(max_key)
    final_dtc.fit(X_mixed, y_mixed)
    ## print("Using Best Parameter Settings Found After Tuning and Mixed Valid/Train Sets: ")
    y_pred = final_dtc.predict(X_test)
    acc = accuracy_score(y_test,y_pred)*100
    f1 = f1_score(y_test, y_pred)
    
    ## print("Accuracy: ", acc)
    ## print("F1 Score: ", f1)
    return acc, f1
```
